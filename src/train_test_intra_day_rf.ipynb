## train a random forest classifier to predict increase/decrease of intraday stock price 

```
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```
#iex_stg2_output_file = 'ALL_TICKERS_STG2_01_OUT.CSV'
iex_stg2_output_file = 'ALL_TICKERS_STG2_OUT.CSV'
dataset = pd.read_csv(iex_stg2_output_file)
columns = list(dataset.columns)
random_state = 42
```

```
dataset.head()
```

```
# perform some EDA
dataset.groupby('target')['price_diff_2_4'].count()
#dataset.describe()
#dataset.groupby('p_sma_23')['target'].count()
```

```
X = dataset.iloc[:, 1:-1].values
y = list(dataset.target)
```

```
columns_to_standardize = columns[1:-1]
dataset[columns_to_standardize] = StandardScaler().fit_transform(dataset[columns_to_standardize])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=random_state)
```

```
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```
rfc=RandomForestClassifier(random_state=random_state)
```

```
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, n_jobs=10, verbose=3, return_train_score=True)
CV_rfc.fit(X_train, y_train)
```
